3.  Compare UI vs. UX
- List three key differences between UI and UX based on your analysis ?
##  **Ans**
    UI (User Interface) and UX (User Experience) are closely related concepts in design, but they focus on different aspects of the user interaction with a product. Here are three key differences between UI and UX based on the analysis of PhonePe:    

1. Focus and Scope
   

UI: UI design is primarily concerned with the visual elements of a product, including layout, colors, typography, icons, and overall aesthetics. It focuses on how the product looks and how users interact with these visual components.    

UX: UX design encompasses the overall experience a user has with a product, including usability, accessibility, and the emotional response during interactions. It focuses on the entire journey a user takes, from discovering the product to completing tasks and achieving goals.    

2. Goals and Objectives
   

UI: The main goal of UI design is to create visually appealing and functional interfaces that facilitate user interactions. It aims to enhance the aesthetic quality of the product and ensure that users can easily navigate through the interface.    

UX: The primary objective of UX design is to optimize the overall user experience by making it as seamless, efficient, and enjoyable as possible. It aims to understand user needs, behaviors, and pain points to create solutions that enhance satisfaction and task completion.
